#The Highest Bank Closure Year

In this analysis I will be visualising the highest years of bank closures from 2000 to 2023.

```python
import pandas as pd
```

```python
banklist = pd.read_csv('../data/raw/banklist.csv', encoding='utf-8')
banklist['Closing Date�'] = pd.to_datetime(banklist['Closing Date�'])
banklist['Year'] = banklist['Closing Date�'].dt.year
aggregated_data = banklist.groupby('Year').agg(count=pd.NamedAgg(column="Year", aggfunc="count")).reset_index()
print(aggregated_data)
```

```python
banklist
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(aggregated_data['Year'])
plt.bar(aggregated_data['Year'], aggregated_data['count'])

# Add titles and labels
plt.title('Bank Failures From 2000 to 2023')
plt.xlabel('Years')
plt.ylabel('Closures')

# Show the plot
plt.show()
```
